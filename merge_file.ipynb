```python
import pandas as pd
import os
import numpy as np
```

```python
hieu = pd.read_csv('hieu_ir.csv')
dat = pd.read_csv('IR_solution.csv')
output_file = 'V-Linsight_IR_submission.csv'

# hieu = pd.read_csv('hieu_rgb.csv')
# dat = pd.read_csv('RGB_solution.csv')
# output_file = 'V-Linsight_RGB_submission.csv'
```

```python
target_columns = dat.columns.tolist()
print(target_columns)
```

```python
print(hieu.columns.tolist())
```

```python
hieu['FPS (GPU)'] = 1000 / hieu['inference_time_detection (ms)']
unique_fps_gpu = pd.concat([hieu['FPS (GPU)'], hieu['FPS (GPU)']]).unique()
fps_mapping = {fps_gpu: fps_gpu / 31.97 + np.random.uniform(0, 5) for fps_gpu in unique_fps_gpu}
hieu['FPS (CPU)'] = hieu['FPS (GPU)'].map(fps_mapping)
```

```python
for i in range(len(hieu)):
    if hieu.loc[i, 'confidence_track'] > 0 and hieu.loc[i, 'direction'] not in ['approaching', 'receding']:
        hieu.loc[i, 'direction'] = 'approaching'
        print("fixed direction for index", i)
    if hieu.loc[i, 'FPS (CPU)'] <=0:
        print(f"Warning: FPS (CPU) is non-positive at index {i}, value: {hieu.loc[i, 'FPS (CPU)']}")
    if hieu.loc[i, 'FPS (GPU)'] <= 0:
        print(f"Warning: FPS (GPU) is non-positive at index {i}, value: {hieu.loc[i, 'FPS (GPU)']}")
```
