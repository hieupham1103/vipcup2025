```python
import ultralytics
from ultralytics import YOLO
import os
```

```python
model_path = "checkpoints/yolov10n/best.pt"
# model_path = "/home/cvpr2025/objects-interpolation/checkpoints/yolov10n.pt"
```

```python
model = YOLO(model_path)
```

```python
video_folder = "/data4t/vipcup/hieu/vipcup_data/Test_detection_tracking/RGB/videos"
videos = [os.path.join(video_folder, f) for f in os.listdir(video_folder) if f.endswith('.mp4')]
```

```python
for video in videos:
  results = model.track(
                      source=video,
                      tracker="config/bytetrack.yml",
                      save=True,
                      device="cuda:1",
                      )
```
